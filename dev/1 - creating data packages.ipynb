```python
import bw2calc
import numpy as np
```

```python
def multiLCA(functional_unit, list_methods):
    
    # Step 1
    lca = bw2calc.LCA(functional_unit)
    lca.lci()
    act_dict = {bw2data.get_activity(k): v for k, v in lca.activity_dict.items()}
    act_dict_rev = {v: k["location"] for k, v in act_dict.items()}
    CH_idx = [k for k, v in act_dict_rev.items() if v=="CH"]
    RoW_idx = [i for i in act_dict_rev if i not in CH_idx]
    
    # Step 2
    C_matrices = {}
    
    # Step 3
    for method in list_methods:
        lca.switch_method(method)
        C_matrices[method] = lca.characterization_matrix
    
    # Step 4
    results_CH, results_RoW = np.empty((len(list_methods))), np.empty((len(list_methods)))
    
    # Step 5
    for method_index, method in enumerate(list_methods):
        r = (C_matrices[method] * lca.inventory)
        results_CH[method_index] = r[:, CH_idx].sum()
        results_RoW[method_index] = r[:, RoW_idx].sum()
            
    return results_CH, results_RoW
```

```python
results = []
for k, v in res.items():
    print(k)
    for tech, data in v.items():
        print(f"-- {tech}")
        r = multiLCA({data["dataset"]: 1}, methods)

        results.extend(
            [
                [
                    k, tech, "CH", " - ".join(methods[m])
                ] + [r[0][m]] for m, _ in enumerate(methods)
            ]
        )

        results.extend(
            [
                [
                    k, tech, "RoW", " - ".join(methods[m])
                ] + [r[1][m]] for m, _ in enumerate(methods)
            ]
        )
```
